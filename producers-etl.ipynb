```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
tickets = pd.read_csv("Resources/AnnualTicketSales.csv")
tickets.head()
```

```python
new_tickets = tickets.drop(columns=['Unnamed: 5'])
new_tickets
new_tickets[['TOTAL BOX OFFICE', 'TICKETS SOLD']] = new_tickets[['TOTAL BOX OFFICE', 'TICKETS SOLD']] .replace('[\$\,\.]', '', regex=True).astype(int)
new_tickets['TOTAL INFLATION ADJUSTED BOX OFFICE'] = new_tickets['TOTAL INFLATION ADJUSTED BOX OFFICE'].replace('[\$\,\.]', '', regex=True).astype(int)
new_tickets['AVERAGE TICKET PRICE'] = new_tickets['AVERAGE TICKET PRICE'].replace('[\$\,\.]', '', regex=True).astype(int)

total_tickets= new_tickets.rename(columns={"YEAR": "year",
                                                  "TICKETS SOLD": "tickets_sold",
                                                  "TOTAL BOX OFFICE": "total_box_office",
                                                  "TOTAL INFLATION ADJUSTED BOX OFFICE": "total_inflation",
                                                  "AVERAGE TICKET PRICE": "average_ticket_price"})

total_tickets
```

```python
highest_g = pd.read_csv("Resources/HighestGrossers.csv")

highest_g[['TOTAL FOR YEAR', 'TOTAL IN 2019 DOLLARS', 'TICKETS SOLD']] = highest_g[['TOTAL FOR YEAR', 'TOTAL IN 2019 DOLLARS', 'TICKETS SOLD']].replace('[\$\,\.]', '', regex=True).astype(int)

highest_gross= highest_g.rename(columns={"YEAR": "year",
                                           "MOVIE" : "movie",
                                           "GENRE" : "genre",
                                           "MPAA RATING": "mpaa_rating",
                                           "DISTRIBUTOR": "distributor",
                                           "TOTAL FOR YEAR": "yearly_total",
                                           "TOTAL IN 2019 DOLLARS": "2019_total",
                                          "TICKETS SOLD": "tickets_sold",})

highest_gross
```

```python
popular = pd.read_csv("Resources/PopularCreativeTypes.csv")
popular.head()

popular_new = popular.drop([0,9])

popular_new[['TOTAL GROSS', 'AVERAGE GROSS']] = popular_new[['TOTAL GROSS', 'AVERAGE GROSS']].replace('[\$\,\.]', '', regex=True).astype(int)

popular_new['MARKET SHARE'] = popular_new['MARKET SHARE'].str.rstrip('%').astype('float') 

popular_types= popular_new.rename(columns={"RANK": "rank",
                                           "MOVIE" : "movie",
                                           "CREATIVE TYPES" : "creative_type",
                                           "MOVIES": "movies",
                                           "TOTAL GROSS": "total_gross",
                                           "AVERAGE GROSS": "average_gross",
                                           "MARKET SHARE" : "market_share"})



popular_types
```

```python
top_dist = pd.read_csv("Resources/TopDistributors.csv")
top_dist.head()


top_dist[['TOTAL GROSS', 'AVERAGE GROSS']] = top_dist[['TOTAL GROSS', 'AVERAGE GROSS']].replace('[\$\,\%]', '', regex=True).astype(int)
top_dist['MARKET SHARE'] = top_dist['MARKET SHARE'].str.rstrip('%').astype('float') 

top_distributors = top_dist.rename(columns={"RANK": "rank",
                                             "DISTRIBUTORS": "distributors",
                                             "MOVIES": "movies",
                                             "TOTAL GROSS": "total_gross",
                                             "AVERAGE GROSS": "average_gross",
                                             "MARKET SHARE": "market_share"})

top_distributors


```

```python
top_genre = pd.read_csv("Resources/TopGenres.csv")

top_genre[['TOTAL GROSS', 'AVERAGE GROSS']] = top_genre[['TOTAL GROSS', 'AVERAGE GROSS']].replace('[\$\,\.]', '', regex=True).astype(int)
top_genre['MARKET SHARE'] = top_genre['MARKET SHARE'].str.rstrip('%').astype('float') 

# Rename the column headers
top_genres = top_genre.rename(columns={"RANK": "rank",
                                       "GENRES": "genres",
                                       "MOVIES": "movies",
                                       "TOTAL GROSS": "total_gross",
                                       "AVERAGE GROSS": "average_gross",
                                       "MARKET SHARE": "market_share"})

top_genres
```

```python
top_gross = pd.read_csv("Resources/TopGrossingRatings.csv")
top_gross[['TOTAL GROSS', 'AVERAGE GROSS']] = top_gross[['TOTAL GROSS', 'AVERAGE GROSS']].replace('[\$\,\.]', '', regex=True).astype(int)
top_gross['MARKET SHARE'] = top_gross['MARKET SHARE'].str.rstrip('%').astype('float') 

top_gross_ratings = top_gross.rename(columns={"RANK": "rank",
                                              "MPAA RATINGS": "mpaa_ratings",
                                              "MOVIES": "movies",
                                              "TOTAL GROSS": "total_gross",
                                              "AVERAGE GROSS": "average_gross",
                                              "MARKET SHARE": "market_share"})

top_gross_ratings
```

```python
top_gross_src = pd.read_csv("Resources/TopGrossingSources.csv")
top_gross_src.head()

top_gross_src[['TOTAL GROSS', 'AVERAGE GROSS']] =top_gross_src[['TOTAL GROSS', 'AVERAGE GROSS']].replace('[\$\,\%]', '', regex=True).astype(int)
top_gross_src['MARKET SHARE'] = top_gross_src['MARKET SHARE'].str.rstrip('%').astype('float') 

top_gross_source= top_gross_src.rename(columns={"RANK": "rank",
                                                         "SOURCES": "sources",
                                                         "MOVIES": "movies",
                                                         "TOTAL GROSS": "total_gross",
                                                       "AVERAGE GROSS": "average_gross", 
                                                       "MARKET SHARE": "market_share"})
top_gross_source
```

```python
top_prod = pd.read_csv("Resources/TopProductionMethods.csv")
top_prod.head()

top_prod[['TOTAL GROSS', 'AVERAGE GROSS']] = top_prod[['TOTAL GROSS', 'AVERAGE GROSS']].replace('[\$\,\%]', '', regex=True).astype(int)
top_prod ['MARKET SHARE'] = top_prod ['MARKET SHARE'].str.rstrip('%').astype('float') 

top_production = top_prod.rename(columns={"RANK": "rank",
                                          "PRODUCTION METHODS": "production_methods",
                                          "MOVIES": "movies",
                                          "TOTAL GROSS": "total_gross",
                                          "AVERAGE GROSS": "average_gross",
                                          "MARKET SHARE": "market_share"})

top_production

```

```python
wide_release = pd.read_csv("Resources/WideReleasesCount.csv")

wide_release_new = wide_release.drop(columns=['Unnamed: 9','TOTAL MAJOR 6'])
wide_release_count = wide_release_new.rename(columns={"YEAR": "year",
                                                      "WARNER BROS": "warner_bros",
                                                      "WALT DISNEY": "walt_disney",
                                                      "20TH CENTURY FOX": "20th_century_fox",
                                                      "PARAMOUNT PICTURES": "paramount_pictures",
                                                      "SONY PICTURES": "sony_pictures","UNIVERSAL": "universal",
                                                      "TOTAL OTHER STUDIOS": "total_other_studios"})

wide_release_count
```

```python
connection_string = "postgres:password@localhost:5432/producers_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
total_tickets.to_sql(name='total_tickets', con=engine, if_exists='append', index=True)
popular_types.to_sql(name='highest_gross', con=engine, if_exists='append', index=True)
top_distributors.to_sql(name='top_distributors', con=engine, if_exists='append', index=True)
top_genres.to_sql(name='top_genres', con=engine, if_exists='append', index=True)
top_gross_ratings.to_sql(name='top_gross_ratings', con=engine, if_exists='append', index=True)
top_gross_source.to_sql(name='top_gross_source', con=engine, if_exists='append', index=True)
top_production.to_sql(name='top_production', con=engine, if_exists='append', index=True)
wide_release_count.to_sql(name='wide_release_count', con=engine, if_exists='append', index=True)
```
